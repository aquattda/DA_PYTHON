```python
import numpy as np
from deap import base, creator, tools, algorithms
from pyswarm import pso

# Các hàm mục tiêu (giữ nguyên mã gốc)
def sphere(individual, for_ga=False):
    individual = np.array(individual)
    result = np.sum(individual**2)
    return (result,) if for_ga else result

def rastrigin(individual, for_ga=False):
    individual = np.array(individual)
    A = 10
    result = A * len(individual) + np.sum(individual**2 - A * np.cos(2 * np.pi * individual))
    return (result,) if for_ga else result

def ackley(individual, for_ga=False):
    individual = np.array(individual)
    n = len(individual)
    a = 20
    b = 0.2
    c = 2 * np.pi
    sum_sq = np.sum(individual**2)
    cos_term = np.sum(np.cos(c * individual))
    result = -a * np.exp(-b * np.sqrt(sum_sq / n)) - np.exp(cos_term / n) + a + np.exp(1)
    return (result,) if for_ga else result

def rosenbrock(individual, for_ga=False):
    individual = np.array(individual)
    result = np.sum(100.0 * (individual[1:] - individual[:-1]**2)**2 + (1 - individual[:-1])**2)
    return (result,) if for_ga else result

# Genetic Algorithm (GA) (giữ nguyên mã gốc)
def ga_optimization(func, generations=100, population_size=50, cxpb=0.7, mutpb=0.2, indpb=0.2):
    if hasattr(creator, "FitnessMin"):
        del creator.FitnessMin
    if hasattr(creator, "Individual"):
        del creator.Individual

    creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMin)

    def create_individual():
        return [np.random.uniform(-5, 5) for _ in range(10)]

    toolbox = base.Toolbox()
    toolbox.register("individual", tools.initIterate, creator.Individual, create_individual)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    toolbox.register("mate", tools.cxBlend, alpha=0.5)
    toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=indpb)
    toolbox.register("select", tools.selTournament, tournsize=3)
    toolbox.register("evaluate", lambda ind: func(ind, for_ga=True))

    population = toolbox.population(n=population_size)
    algorithms.eaSimple(population, toolbox, cxpb=cxpb, mutpb=mutpb, ngen=generations,
                        stats=None, halloffame=None, verbose=False)

    best_individual = tools.selBest(population, 1)[0]
    return best_individual.fitness.values[0]

# Particle Swarm Optimization (PSO) (giữ nguyên mã gốc)
def pso_optimization(func):
    lb = [-5] * 10
    ub = [5] * 10
    best_pos, best_val = pso(func, lb, ub, swarmsize=50, maxiter=100)
    return best_val

# SELOA Optimization
def seloa_optimization(func, population_size=50, generations=100, alpha=0.5, beta=0.3):
    """
    Triển khai SELOA: Socio Evolution Learning Optimization Algorithm.
    - population_size: Kích thước quần thể
    - generations: Số thế hệ
    - alpha: Hệ số học hỏi từ global best
    - beta: Hệ số học hỏi từ cá thể ngẫu nhiên
    """
    # Khởi tạo quần thể
    dim = 10  # 10 chiều
    population = [np.random.uniform(-5, 5, dim) for _ in range(population_size)]
    fitness = [func(ind) for ind in population]

    # Tìm global best ban đầu
    best_idx = np.argmin(fitness)
    global_best = population[best_idx].copy()
    global_best_fitness = fitness[best_idx]

    # Vòng lặp chính
    for _ in range(generations):
        for i in range(population_size):
            # Chọn ngẫu nhiên một cá thể khác để học hỏi
            random_idx = np.random.randint(0, population_size)
            while random_idx == i:
                random_idx = np.random.randint(0, population_size)

            # Cập nhật vị trí cá thể
            new_position = population[i].copy()
            for j in range(dim):
                # Học từ global best và cá thể ngẫu nhiên
                learn_from_best = alpha * np.random.random() * (global_best[j] - population[i][j])
                learn_from_random = beta * np.random.random() * (population[random_idx][j] - population[i][j])
                # Thêm nhiễu ngẫu nhiên để khám phá
                noise = np.random.uniform(-0.1, 0.1)
                new_position[j] += learn_from_best + learn_from_random + noise

            # Giới hạn vị trí trong [-5, 5]
            new_position = np.clip(new_position, -5, 5)

            # Đánh giá fitness mới
            new_fitness = func(new_position)

            # Cập nhật cá thể nếu fitness mới tốt hơn
            if new_fitness < fitness[i]:
                population[i] = new_position
                fitness[i] = new_fitness

                # Cập nhật global best nếu cần
                if new_fitness < global_best_fitness:
                    global_best = new_position.copy()
                    global_best_fitness = new_fitness

    return global_best_fitness

# Hàm đánh giá các thuật toán (đã sửa để bao gồm SELOA)
def evaluate_algorithms():
    functions = [sphere, rastrigin, ackley, rosenbrock]
    algorithms = {
        'GA': ga_optimization,
        'PSO': pso_optimization,
        'SELOA': seloa_optimization
    }

    results = {}
    for func in functions:
        func_name = func.__name__
        results[func_name] = {}
        for algo_name, algo_func in algorithms.items():
            algo_results = []
            for _ in range(10):  # Chạy mỗi thuật toán 10 lần
                algo_results.append(algo_func(func))
            results[func_name][algo_name] = np.mean(algo_results)

    return results

# In kết quả
if __name__ == "__main__":
    results = evaluate_algorithms()
    for func, result in results.items():
        print(f"Hàm: {func}")
        for algo, value in result.items():
            print(f"  {algo}: {value:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
POPULATION_SIZE = 50
GENERATIONS = 100
MUTATION_RATE = 0.1
LEARNING_RATE = 0.2
TASK_DIFFICULTY = 1.0

# Initialize population with random strategies (weights between 0 and 1)
population = np.random.rand(POPULATION_SIZE, 3)  # 3D strategy vector per agent
fitness_history = []

def evaluate_fitness(individual):
    """Evaluate fitness based on a cooperative task (quadratic reward function)."""
    return np.sum(individual) - TASK_DIFFICULTY * np.sum(individual**2)

def social_learning(population, fitnesses):
    """Agents learn from better-performing peers."""
    new_population = population.copy()
    for i in range(POPULATION_SIZE):
        # Find a random peer
        peer_idx = np.random.randint(POPULATION_SIZE)
        if fitnesses[peer_idx] > fitnesses[i]:
            # Learn: move strategy toward peer's strategy
            new_population[i] += LEARNING_RATE * (population[peer_idx] - population[i])
    return np.clip(new_population, 0, 1)

def selection(population, fitnesses):
    """Select individuals based on fitness (tournament selection)."""
    selected = []
    for _ in range(POPULATION_SIZE):
        idx1, idx2 = np.random.randint(0, POPULATION_SIZE, 2)
        selected.append(population[idx1] if fitnesses[idx1] > fitnesses[idx2] else population[idx2])
    return np.array(selected)

def crossover(parent1, parent2):
    """Perform crossover between two parents."""
    if np.random.rand() < 0.7:  # Crossover probability
        point = np.random.randint(1, len(parent1))
        child = np.concatenate((parent1[:point], parent2[point:]))
        return child
    return parent1.copy()

def mutation(individual):
    """Apply mutation to an individual."""
    for i in range(len(individual)):
        if np.random.rand() < MUTATION_RATE:
            individual[i] += np.random.normal(0, 0.1)
    return np.clip(individual, 0, 1)

# Main evolution loop
for gen in range(GENERATIONS):
    # Evaluate fitness
    fitnesses = np.array([evaluate_fitness(ind) for ind in population])
    fitness_history.append(np.mean(fitnesses))

    # Social learning
    population = social_learning(population, fitnesses)

    # Evolutionary operations
    new_population = []
    selected = selection(population, fitnesses)

    for i in range(0, POPULATION_SIZE, 2):
        parent1 = selected[i]
        parent2 = selected[min(i+1, POPULATION_SIZE-1)]
        child1 = crossover(parent1, parent2)
        child2 = crossover(parent2, parent1)
        new_population.append(mutation(child1))
        new_population.append(mutation(child2))

    population = np.array(new_population[:POPULATION_SIZE])

# Plot the average fitness over generations
plt.figure(figsize=(8, 6))
plt.plot(range(GENERATIONS), fitness_history, 'b-', label='Average Fitness')
plt.title('Socio Evolution and Learning Optimization')
plt.xlabel('Generation')
plt.ylabel('Average Fitness')
plt.grid(True)
plt.legend()
plt.savefig('socio_evolution_plot.png')
```
